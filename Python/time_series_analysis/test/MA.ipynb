```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("../../../")
from Python.time_series_analysis import white_noise
```

```python
theta = [-0.9, -0.5, 0.5, 0.9]
sigma = [0.5, 2]
c = 0
sample = 100

fig = plt.figure(figsize=(15, 5))
fig.subplots_adjust(hspace=0.5, wspace=0.2)

for k in range(len(sigma)):
    for j in range(len(theta)):

        noise = white_noise(dlen=sample + 1, mean = 0, std=sigma[k])
        y0 = 0  # 時刻0のやつ
        y = np.zeros(sample + 1)
        y[0] = y0

        for i in range(sample):
            y[i + 1] = c + noise[i + 1] + noise[i] * theta[j]
        
        ax = fig.add_subplot(2, 4, 1 + j + 4 * k)
        ax.set_title("c = {0}, Θ = {1}, ε = {2}".format(c, theta[j], sigma[k]))
        ax.plot([int(i) for i in range(sample + 1)], y)
plt.show()
```
