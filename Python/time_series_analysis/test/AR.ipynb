```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("../../../")
from Python.time_series_analysis import white_noise
```

```python
phi = [-1.1, -0.8, -0.2, 0.2, 0.8, 1.1]
sigma = [0.5, 2]
center = 0
sample = 100
fig = plt.figure(figsize=(16, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.4)

for k in range(len(sigma)):
    for j in range(len(phi)):
        noise = white_noise(dlen=sample + 1, mean=0, std=sigma[k])
        y0 = 0
        y = np.zeros(sample + 1)
        y[0] = y0

        for i in range(sample):
            y[i + 1] = center + phi[j] * y[i] + noise[i]

        ax = fig.add_subplot(2, 6, 1 + j + 6 * k)
        ax.set_title("c = {0}, Φ = {1}, ε = {2}".format(center, phi[j], sigma[k]))
        ax.plot([int(i) for i in range(1, sample + 2)], y)

plt.show
```
